```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(y,t):
    dydt = (1 - t * y * y) / (t * t * y - 1)
    return dydt

# initial condition
y0 = 0

# time points
t = np.linspace(0,1)

# solve ODE
y = odeint(model,y0,t)

# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```
