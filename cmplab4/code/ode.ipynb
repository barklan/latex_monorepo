```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format = 'retina'
sns.set(font_scale=0.6, rc={"figure.figsize":(3, 2)})
```

### Gleb

```python
def model(y, t):
    dydt = (1 - t * y * y) / (t * t * y - 1)
    return dydt


y0 = 0
t = np.linspace(0, 1)
y = odeint(model, y0, t)
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.show()
```

### Ilyana

```python
def model(y, t):
    dydt = ((t ** 4) * y * y - t * t * y - 20) / (t ** 3)
    return dydt


y0 = 4
t = np.linspace(1, 2)
y = odeint(model, y0, t)
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.show()
```
